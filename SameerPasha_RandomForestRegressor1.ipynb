```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\NV Studios i7\Desktop\ATS\AI&ML\DataSets\Diamonds Prices2022.csv")
df.head()
```

```python
df.drop(['S.No',	'carat','color',	'clarity',	'depth',	'table','x',	'y',	'z'],axis=1,inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['cut']=le.fit_transform(df['cut'])
df.head()
```

```python
x=df.drop('price',axis='columns')
x.head()
```

```python
y=df['price']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
x_train.shape
```

```python
y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
reg=RandomForestRegressor(n_estimators=50,random_state=5)
```

```python
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
```

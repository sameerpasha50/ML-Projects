```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\NV Studios i7\Desktop\ATS\AI&ML\DataSets\USA Housing Dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['date','bathrooms',	'sqft_living',	'sqft_lot','waterfront',	'view',	'condition',	'sqft_above',	'sqft_basement',	'yr_built',	'yr_renovated',	'street',	'city',	'statezip',	'country'],axis=1,inplace=True)
df.head()
```

```python
x=df.drop('price',axis='columns')
x.head()
```

```python
y=df['price']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
x_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor,plot_tree
reg=DecisionTreeRegressor(random_state=1)
reg.fit(x_train,y_train)
```

```python
y_pred=reg.predict(x_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(50,50))
plot_tree(reg,feature_names=x.columns,filled=True)
plt.show()
```

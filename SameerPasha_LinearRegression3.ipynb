```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\NV Studios i7\Desktop\ATS\AI&ML\DataSets\index_1.csv")
df.head()
```

```python
df.drop(['date','datetime','card'],axis='columns',inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cols_to_en=['cash_type','coffee_name']
for cols in cols_to_en:
    df[cols]=le.fit_transform(df[cols])
df.head()
```

```python
x=df.drop(['money'],axis='columns')
x.head()
```

```python
y=df['money']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
x_train.shape
```

```python
y_test.shap
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\NV Studios i7\Desktop\ATS\AI&ML\DataSets\laptop_prices.csv")
df.head()
```

```python
df.drop(['Product',	'TypeName',	'Inches',	'Ram','Weight','Screen','ScreenW',	'ScreenH','CPU_freq',	'CPU_model',	'PrimaryStorage',	'SecondaryStorage','GPU_model'],axis=1,inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cols_to_encode=['Company',	'OS','Touchscreen',	'IPSpanel',	'RetinaDisplay',	'CPU_company'	,'PrimaryStorageType',	'SecondaryStorageType',	'GPU_company']
for cols in cols_to_encode:
    df[cols]=le.fit_transform(df[cols])
df.head()
```

```python
x=df.drop(['Price_euros'],axis=1)
x.head()
```

```python
y=df['Price_euros']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
x_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\NV Studios i7\Desktop\ATS\AI&ML\DataSets\Students Social Media Addiction.csv")
```

```python
df.head()
```

```python
df.drop(['Student_ID',	'Age','Country',	'Avg_Daily_Usage_Hours',	'Most_Used_Platform','Sleep_Hours_Per_Night',	'Mental_Health_Score','Conflicts_Over_Social_Media',	'Addicted_Score'],axis=1,inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cols_to_encode=['Gender'	,'Academic_Level',	'Affects_Academic_Performance',	'Relationship_Status']
for cols in cols_to_encode:
    df[cols]=le.fit_transform(df[cols])
df.head()
```

```python
x=df.drop('Affects_Academic_Performance',axis='columns')
x.head()
```

```python
y=df['Affects_Academic_Performance']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
x_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
clf = DecisionTreeClassifier(random_state=1)
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(100,200))
plot_tree(clf,feature_names = x.columns,filled=True)
```
